```python
# Plotting Charts
def create_plot(vector, data_name_short, title):
    df = sc.vectors_to_df([vector])
    #today = datetime.datetime.today().isoformat()
    
    #df = ticker_data.history(period='1d', start='2017-07-15', end=today[:10])
    plt.subplots(figsize=(20,10))
    plt.plot(df[-100:], label=data_name_short, color='blue')
    #plt.fill(df[-100:], color='red')
    plt.margins(x=0.01,tight=True)
    plt.xlabel("Date", fontdict={'fontsize':'14'})
    plt.ylabel(data_name_short, fontdict={'fontsize':'14'})
    plt.title(title, fontdict={'fontsize':'18'})
    plt.xticks(size = 14)
    plt.yticks(size = 14)
    plt.savefig(f'Images/{data_name_short}.png', bbox_inches="tight")
    
create_plot("v2062815", "Unemployment Rate", "Canada Seasonally Adjusted Estimated Unemployment Rate (15 years and over)" )
create_plot("v2062816", "Participation Rate", "Canada Seasonally Adjusted Estimated Participation Rate (15 years and over)")
create_plot("v86822775", "Budget Balance", "Canada Budget Deficit/Surplus Monthly")
create_plot("v2063948", "Ontario Unemployment", "Unemployment in Ontario Seasonally Adjusted")
```
