### PBOC Open Market Operations Data Analysis

```python
# PBOC Open Market Operations
macro_china_gksccz = ak.macro_china_gksccz()
macro_china_pboc_omo_df = pd.DataFrame(macro_china_gksccz)
macro_china_pboc_omo_df
```

```python
# Count number of Reverse repo vs. repo operations
macro_china_pboc_omo_df.trading_method.value_counts()

# Create datetime index
macro_china_pboc_omo_df["Date"] = pd.to_datetime(macro_china_pboc_omo_df["operation_from_date"])
macro_china_pboc_omo_df = macro_china_pboc_omo_df.set_index(["Date"]) 
macro_china_pboc_omo_df = macro_china_pboc_omo_df.drop(columns=["operation_from_date"])
```

```python
# Change dtypes to 'float'
macro_china_pboc_omo_df["deal_amount"] = pd.to_numeric(macro_china_pboc_omo_df["deal_amount"], downcast="float")
macro_china_pboc_omo_df["rate"] = pd.to_numeric(macro_china_pboc_omo_df["rate"], downcast="float")
macro_china_pboc_omo_df.dtypes


# Groupby Trading Method
# grouped_df = macro_china_pboc_omo_df.groupby([macro_china_pboc_omo_df.index, "trading_method"]).sum()
# grouped_df.tail()
```

```python
# Creating two dataframes based on trading_method
reverse_repo_type = macro_china_pboc_omo_df.trading_method == "逆回购"
positive_repo_type = macro_china_pboc_omo_df.trading_method == "正回购"

reverse_repo_df = macro_china_pboc_omo_df[reverse_repo_type]
positive_repo_df = macro_china_pboc_omo_df[positive_repo_type]
```

```python
# Resampling Data to get monthly sums

reverse_repo_df_resampled = reverse_repo_df.deal_amount.resample('M').sum()
reverse_repo_df_resampled = pd.DataFrame(reverse_repo_df_resampled)
reverse_repo_df_resampled.tail()

positive_repo_df_resampled = positive_repo_df.deal_amount.resample('M').sum()
positive_repo_df_resampled = pd.DataFrame(positive_repo_df_resampled)
positive_repo_df_resampled.tail()
```

```python
# Concat separate dataframes
positive_repo_df_resampled = positive_repo_df_resampled.rename(columns={"deal_amount": "Positive_Repo_Amount"})
reverse_repo_df_resampled = reverse_repo_df_resampled.rename(columns={"deal_amount": "Reverse_Repo_Amount"})
master_repo_df = pd.concat([positive_repo_df_resampled, reverse_repo_df_resampled], axis="columns", join="outer")
master_repo_df.index
```

```python
# Plot the China Open Market Operations Data

x = reverse_repo_df_resampled.index
y = reverse_repo_df_resampled["Reverse_Repo_Amount"]
x2 = positive_repo_df_resampled.index
y2 = positive_repo_df_resampled["Positive_Repo_Amount"]
plt.subplots(figsize=(15,8))
plt.bar(x, y, label="Reverse Repo - 逆回购",width=5)
plt.bar(x2, y2, label="Repo - 正回购",width=5)
plt.legend(prop={'size': 16})
plt.xlabel("Date")
plt.ylabel("Transaction Volume(summed by month) (CNY 100 Million)")
plt.title(f"PBOC Open Market Operations - 央行公开市场操作", fontdict={'fontweight':'bold', 'fontsize':'22'})
plt.savefig(f'Images/China/PBOC Open Market Operations Total.png', bbox_inches="tight")
plt.show()
```

```python
# Plot the China Open Market Operations Data
x = positive_repo_df_resampled.index
y = positive_repo_df_resampled["Positive_Repo_Amount"]
plt.subplots(figsize=(15,8))
plt.bar(x,y, label="正回购",width=5)
plt.legend(prop={'size': 16})
plt.xlabel("Date")
plt.ylabel("Transaction Volume(summed by month) (CNY 100 Million)")
plt.title(f"PBOC Open Market Operations - Repo Agreement - 央行公开市场操作 - 正回购 ", fontdict={'fontweight':'bold', 'fontsize':'22'})
plt.savefig(f'Images/China/PBOC Repo Operations.png', bbox_inches="tight")
plt.show()

```

```python
# Plot the China Open Market Operations Data
x = reverse_repo_df_resampled.index
y = reverse_repo_df_resampled["Reverse_Repo_Amount"]
plt.subplots(figsize=(15,8))
plt.bar(x,y, label="逆回购",width=5)
plt.legend(prop={'size': 16})
plt.xlabel("Date")
plt.ylabel("Transaction Volume(summed by month) (CNY 100 Million)")
plt.title(f"PBOC Open Market Operations - Reverse-Repo Agreement - 央行公开市场操作 - 逆回购", fontdict={'fontweight':'bold', 'fontsize':'22'})
plt.savefig(f'Images/China/PBOC Reverse Repo Operations.png', bbox_inches="tight")
plt.show()
```

### China Lending Benchmark Loan Prime Rate (LPR) Analysis - 中国-利率-贷款报价利率

The LPR is a lending reference rate set monthly by 18 banks. LPR, since August 2019, is loosely pegged to the MLF rate. PBOC however has decreased the use of the MLF loan facility.
Most new and outstanding loans are based on the LPR, while the five-year rate influences the pricing of mortgages.

```python
# China LPR data
# Rate1 = 短期贷款利率: 6个月至1年(含)(%)
# Rate2 = 中长期贷款利率: 5年以上(%)
macro_china_lpr = ak.macro_china_lpr()
macro_china_lpr_df = pd.DataFrame(macro_china_lpr)
macro_china_lpr_df.head()
```

```python
# Plotting China LPR data
labels = macro_china_lpr_df.columns
plt.subplots(figsize=(20,10))
plt.plot(macro_china_lpr_df)
plt.legend(labels, prop={'size': 16})
plt.xlabel("Date")
plt.ylabel("Rate %")
plt.title(f"China Lending Benchmark Loan Prime Rate (LPR) - 中国-利率-贷款报价利率", fontdict={'fontweight':'bold', 'fontsize':'22'})
plt.savefig(f'Images/China/China Lending Benchmark Loan Prime Rate_LPR.png', bbox_inches="tight")
plt.show()
```

```python
# Plotting China LPR data - last 5 years
labels = macro_china_lpr_df.columns
df = macro_china_lpr_df[:'2015']
plt.subplots(figsize=(20,10))
plt.plot(df)
plt.legend(labels, prop={'size': 16})
plt.xlabel("Date")
plt.ylabel("Rate %")
plt.title(f"China Lending Benchmark Loan Prime Rate (LPR) - 中国-利率-贷款报价利率", fontdict={'fontweight':'bold', 'fontsize':'22'})
plt.savefig(f'Images/China/China Lending Benchmark Loan Prime Rate_LPR last 5 years.png', bbox_inches="tight")
plt.show()
```

### SHIBOR Analysis

上海银行间同业拆放利率（Shanghai Interbank Offered Rate，简称Shibor）- 以位于上海的全国银行间同业拆借中心为技术平台计算、发布并命名，是由信用等级较高的银行组成报价团自主报出的人民币同业拆出利率计算确定的算术平均利率，是单利、无担保、批发性利率。目前，对社会公布的Shibor品种包括隔夜、1周、2周、1个月、3个月、6个月、9个月及1年。
Shibor报价银行团现由18家商业银行组成。报价银行是公开市场一级交易商或外汇市场做市商，在中国货币市场上人民币交易相对活跃、信息披露比较充分的银行。目前，Shibor与货币市场发展已经形成了良性互动的格局。Shibor在市场化产品定价中得到广泛运用。一是Shibor对债券产品定价的指导性持续增强。二是以Shibor为基准的金融创新产品成交活跃。三是票据转贴现、回购业务初步建立了以Shibor为基准的市场化定价机制。
全国银行间同业拆借中心授权Shibor的报价计算和信息发布。每个交易日根据各报价行的报价，剔除最高、最低各4家报价，对其余报价进行算术平均计算后，得出每一期限品种的Shibor，并于11:00对外发布。

```python
# China Shibor
macro_china_shibor_all = ak.macro_china_shibor_all()
macro_china_shibor_all_df = pd.DataFrame(macro_china_shibor_all)
macro_china_shibor_all_df.head(15)
```

```python
macro_china_shibor_all_df.index = pd.to_datetime(macro_china_shibor_all_df.index)
macro_china_shibor_all_df.index
macro_china_shibor_all_df = macro_china_shibor_all_df.drop(macro_china_shibor_all_df.index[0])
```

```python
# Plotting SHIBOR
SHIBOR_ON_df = macro_china_shibor_all_df[["O/N_定价"]]
SHIBOR_1W_df = macro_china_shibor_all_df[["1W_定价"]]
SHIBOR_1M_df = macro_china_shibor_all_df[["1M_定价"]]

plt.subplots(figsize=(20,10))
#plt.plot(SHIBOR_ON_df[-400:], label="SHIBOR_Overnight_定价")
plt.plot(SHIBOR_1W_df[-400:], label="SHIBOR_1_Week_定价")
plt.plot(SHIBOR_1M_df[-400:], label="SHIBOR_1_Month_定价")
plt.plot()
plt.legend(prop={'size': 16})
plt.xlabel("Date")
plt.ylabel("Rate %")
plt.title(f"上海银行间同业拆放利率（Shanghai Interbank Offered Rate，简称Shibor）", fontdict={'fontweight':'bold', 'fontsize':'22'})
plt.savefig(f'Images/China/SHIBOR_Analysis.png', bbox_inches="tight")
plt.show()
```
